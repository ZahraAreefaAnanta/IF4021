````python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load audio files (replace with your actual file paths)
audio_files = [
    "/content/drive/My Drive/ho_Multimedia/murmullos-7133.mp3",  # Audio dengan volume pelan
    "/content/drive/My Drive/ho_Multimedia/crowd-murmuring-60721.mp3",  # Audio dengan volume sedang
    "/content/drive/My Drive/ho_Multimedia/male-scream-in-fear-123079.mp3",  # Audio dengan volume tinggi
    "/content/drive/My Drive/ho_Multimedia/gyddc-w8lxq.mp3",  # Audio dengan brbrbrbrb
    "/content/drive/My Drive/ho_Multimedia/dbjhd-d3wvn.mp3"  # Audio dengan ssssssss
]

# Initialize lists to hold the loaded audio data and sampling rates
audio_data = []
sampling_rates = []

# Load each audio file using librosa
for file in audio_files:
    audio, sr = librosa.load(file, sr=None)
    audio_data.append(audio)
    sampling_rates.append(sr)

# Function to plot waveform and spectrogram
def plot_waveform_and_spectrogram(audio, sr, title):
    plt.figure(figsize=(14, 8))

    # Plot the waveform
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(audio, sr=sr)
    plt.title(f"Waveform - {title}")
    plt.xlabel("Time (seconds)")
    plt.ylabel("Amplitude")

    # Plot the spectrogram
    plt.subplot(2, 1, 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.title(f"Spectrogram - {title}")
    plt.colorbar(format='%+2.0f dB')
    plt.xlabel("Time (seconds)")
    plt.ylabel("Frequency (Hz)")

    plt.tight_layout()
    plt.show()

# Plot each audio separately with individual titles
titles = ['Quiet Audio', 'Medium Audio', 'Loud Audio', 'brrbrbrbrb', 'esssss']

# Iterate over all audio files
for i in range(len(audio_files)):
    print(f"Visualizing: {titles[i]}")
    plot_waveform_and_spectrogram(audio_data[i], sampling_rates[i], titles[i])
````

````python
import matplotlib.pyplot as plt
import numpy as np
import wave
import os
````

````python
# Fungsi untuk memuat audio dari file WAV
def load_audio(file_path):
    with wave.open(file_path, 'r') as wav_file:
        n_channels = wav_file.getnchannels()
        sample_width = wav_file.getsampwidth()
        framerate = wav_file.getframerate()
        n_frames = wav_file.getnframes()

        audio_content = wav_file.readframes(n_frames)
        audio_data = np.frombuffer(audio_content, dtype=np.int16)

    return audio_data, framerate, sample_width, n_channels
````

````python
# Fungsi untuk membuat exponential fading curve
def exponential_fade(duration_samples, start_value=1e-4, end_value=1.0):
    return np.logspace(np.log10(start_value), np.log10(end_value), duration_samples)
````

````python
# Load file audio
file_path = os.path.join(os.getcwd(), '/content/drive/My Drive/ho_Multimedia', 'Avatar the Last Airbender - Instrumental.wav')

# Cek apakah file ada
if not os.path.exists(file_path):
    print('File tidak ditemukan')
    exit()

audio_data, fs, sample_width, n_channels = load_audio(file_path)
````

````python
# Durasi fading (dalam detik)
fade_duration = 5  # detik
fade_samples = int(fade_duration * fs)

# Fading di awal
fade_in_curve = exponential_fade(fade_samples)

# Fading di akhir
fade_out_curve = exponential_fade(fade_samples)[::-1]
````

````python
# Terapkan fading ke audio
audio_data_fading = audio_data.copy()

# Fading in di awal
audio_data_fading[:fade_samples] = audio_data_fading[:fade_samples] * fade_in_curve

# Fading out di akhir
audio_data_fading[-fade_samples:] = audio_data_fading[-fade_samples:] * fade_out_curve
````

````python
# Buat time axis
time_axis = np.linspace(0, len(audio_data) / fs, num=len(audio_data))

# Visualisasi hasil
plt.figure(figsize=(14, 6))

# Plot original
plt.subplot(2, 1, 1)
plt.plot(time_axis, audio_data, label='Original Audio')
plt.title('Original Audio Waveform')

# Plot audio dengan fading
plt.subplot(2, 1, 2)
plt.plot(time_axis, audio_data_fading, label='Audio with Exponential Fading', color='r')
plt.title('Audio with Exponential Fade In and Fade Out')

plt.tight_layout()
plt.show()
````

Dengan menggunakan file audio yang anda rekam sendiri (suara anda), lakukanlah kompresi dan normalisasi hingga loudness LUFS mencapai kira-kira sekitar -14 LUFS. Berikan penjelasan langkah-langkah yang anda lakukan untuk menyelesaikan tugas ini.

````python
import numpy as np
import wave
import os
import matplotlib.pyplot as plt
````

````python
def load_audio(filepath):
    with wave.open(filepath, 'rb') as wav_file:
        n_channels = wav_file.getnchannels()
        sampwidth = wav_file.getsampwidth()
        framerate = wav_file.getframerate()
        n_frames = wav_file.getnframes()
        audio_content = wav_file.readframes(n_frames)
        audio_data = np.frombuffer(audio_content, dtype=np.int16)
        return audio_data, framerate, n_channels, sampwidth
````

Fungsi ini memuat file audio dari lokasi yang diberikan (filepath) dan mengekstrak informasi dasar seperti jumlah channel, lebar sampel, frekuensi sampel (sampling rate), dan jumlah frame audio.

````python
def normalize_audio(audio_data):
    peak_amplitude = np.max(np.abs(audio_data))
    normalized_audio = (audio_data / peak_amplitude) * np.iinfo(np.int16).max
    return normalized_audio.astype(np.int16)
````

Melakukan normalisasi pada audio sehingga puncak amplitudo (peak amplitude) berada pada nilai maksimal untuk data tipe int16.

````python
def apply_compression(audio_data):
    compressed_audio = np.where(audio_data > 10000, audio_data * 0.7, audio_data)
    compressed_audio = np.where(audio_data < -10000, audio_data * 0.7, compressed_audio)
    return compressed_audio
````

Menerapkan kompresi sederhana pada audio dengan mengurangi amplitudo untuk bagian sinyal yang lebih keras.

````python
def calculate_lufs(audio_data):
    audio_float = audio_data.astype(np.float32) / np.iinfo(np.int16).max
    ms = np.mean(np.square(audio_float))
    lufs = -0.691 + 10 * np.log10(ms)
    return lufs
````

Menghitung LUFS (Loudness Units Full Scale), yang merupakan metrik loudness yang lebih sesuai dengan bagaimana manusia mendengar loudness daripada hanya menggunakan amplitudo puncak.

````python
def save_audio(filepath, audio_data, framerate, n_channels, sampwidth):
    with wave.open(filepath, 'wb') as wav_file:
        wav_file.setnchannels(n_channels)
        wav_file.setsampwidth(sampwidth)
        wav_file.setframerate(framerate)
        wav_file.writeframes(audio_data.tobytes())
````

````python
normalized_audio = normalize_audio(audio_data)
````

Normalisasi awal dilakukan untuk memastikan puncak amplitudo mencapai batas maksimal.

````python
compressed_audio = apply_compression(normalized_audio)
````

Mengurangi dinamika audio dengan menerapkan kompresi sederhana pada amplitudo yang lebih keras.

````python
lufs_before = calculate_lufs(compressed_audio)
````

Menghitung LUFS dari audio yang sudah dikompresi untuk mengetahui loudness sebelum penyesuaian lebih lanjut.

````python
target_lufs = -14
lufs_difference = target_lufs - lufs_before
adjustment_factor = 10**(lufs_difference / 20)
final_audio = compressed_audio * adjustment_factor
final_audio = np.clip(final_audio, -np.iinfo(np.int16).max, np.iinfo(np.int16).max).astype(np.int16)
````

Melakukan penyesuaian volume untuk mengurangi atau meningkatkan loudness hingga mendekati -14 LUFS.

````python
lufs_after = calculate_lufs(final_audio)

# Menampilkan hasil LUFS
print(f"LUFS setelah penyesuaian: {lufs_after}")
````

Rekamlah sebuah audio dengan menggunakan handphone / laptop anda. Cara anda merekam haruslah sedikit unik:

- Lakukanlah perekaman di dekat sumber noise statis seperti kipas angin, AC, atau kipas laptop anda (atau apapun yang merupakan noise frekuensi tinggi)
- Lakukanlah equalisasi pada audio tersebut dengan menggunakan bandpass. Temukan frekuensi cutoff untuk bandpass yang paling sesuai dengan karakteristik audio yang anda rekam.
- Visualisasikan spektrum frekuensi dari audio sebelum di filter dan setelah di filter (dengan ketiga filter yang telah anda buat).

````python
import numpy as np
import matplotlib.pyplot as plt
import wave
import os
from scipy.signal import butter, lfilter

# Fungsi untuk memuat audio
def load_audio(file_path):
    with wave.open(file_path, 'r') as wav_file:
        n_channels = wav_file.getnchannels()
        sample_width = wav_file.getsampwidth()
        fs = wav_file.getframerate()
        n_samples = wav_file.getnframes()
        audio_content = wav_file.readframes(n_samples)
        audio_data = np.frombuffer(audio_content, dtype=np.int16)

    return audio_data, fs, sample_width, n_channels

# Fungsi untuk menerapkan bandpass filter
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyquist = 0.5 * fs
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(order, [low, high], btype='band')
    return b, a

def bandpass_filter(data, lowcut, highcut, fs, order=5):
    b, a = butter_bandpass(lowcut, highcut, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Fungsi untuk menampilkan spektrum frekuensi
def plot_spectrum(audio_data, fs, title):
    n = len(audio_data)
    freqs = np.fft.rfftfreq(n, d=1/fs)
    fft_values = np.fft.rfft(audio_data)

    plt.figure(figsize=(10, 5))
    plt.plot(freqs, np.abs(fft_values))
    plt.title(title)
    plt.xlabel('Frekuensi (Hz)')
    plt.ylabel('Amplitudo')
    plt.xlim(0, fs/2)  # Hanya menampilkan setengah spektrum
    plt.grid()
    plt.show()

# Muat audio yang telah direkam
file_path = '/content/drive/My Drive/ho_Multimedia/audio_3_4.wav'  # Ganti dengan path audio Anda
audio_data, fs, sample_width, n_channels = load_audio(file_path)

# Visualisasikan spektrum frekuensi sebelum filter
plot_spectrum(audio_data, fs, 'Spektrum Frekuensi Sebelum Filter')

# Terapkan bandpass filter (300 Hz - 3400 Hz)
lowcut = 300.0
highcut = 3400.0
filtered_audio = bandpass_filter(audio_data.astype(np.float32), lowcut, highcut, fs)

# Visualisasikan spektrum frekuensi setelah filter
plot_spectrum(filtered_audio, fs, 'Spektrum Frekuensi Setelah Filter')

# Simpan audio yang telah difilter
output_path = 'filtered_audio.wav'  # Ganti dengan path output yang diinginkan
with wave.open(output_path, 'w') as wav_file:
    wav_file.setnchannels(n_channels)
    wav_file.setsampwidth(sample_width)
    wav_file.setframerate(fs)
    wav_file.writeframes(filtered_audio.astype(np.int16).tobytes())

print("Audio telah difilter dan disimpan sebagai:", output_path)
````

Tanyakan pada AI/LLM bagaimana cara membuat noise gate pada audio. Lalu implementasikan noise gate ini pada audio yang telah anda rekam. Jangan lupa copy/paste hasil percakapan anda dengan AI LLM ke notebook anda. Gunakan file audio dari soal nomor 4.

````python
import numpy as np
import wave
import os
import matplotlib.pyplot as plt

# Fungsi untuk memuat audio
def load_audio(file_path):
    with wave.open(file_path, 'r') as wav_file:
        n_channels = wav_file.getnchannels()
        sample_width = wav_file.getsampwidth()
        fs = wav_file.getframerate()
        n_samples = wav_file.getnframes()
        audio_content = wav_file.readframes(n_samples)
        audio_data = np.frombuffer(audio_content, dtype=np.int16)

    return audio_data, fs, sample_width, n_channels

# Fungsi untuk menerapkan noise gate
def noise_gate(audio_data, threshold, ratio, attack_time, release_time, fs):
    output = np.zeros_like(audio_data)
    gain = 1.0

    attack_samples = int(attack_time * fs)
    release_samples = int(release_time * fs)

    for i in range(len(audio_data)):
        if audio_data[i] > threshold:
            gain = 1.0  # Suara di atas threshold, lewatkan tanpa perubahan
        else:
            gain = 0.0  # Suara di bawah threshold, dibungkam

        output[i] = audio_data[i] * gain

        # Apply attack and release
        if i > 0:
            if gain > 0:
                # Increase gain during attack
                gain = min(gain + (1 / attack_samples), 1.0)
            else:
                # Decrease gain during release
                gain = max(gain - (1 / release_samples), 0.0)

        output[i] = audio_data[i] * gain

    return output

# Muat audio yang telah direkam
file_path = '/content/drive/My Drive/ho_Multimedia/audio_3_4.wav'  # Ganti dengan path audio Anda
audio_data, fs, sample_width, n_channels = load_audio(file_path)

# Parameter Noise Gate
threshold = 500  # Ambang batas (ubah sesuai kebutuhan)
ratio = 4.0      # Rasio
attack_time = 0.05  # Waktu serangan dalam detik
release_time = 0.1  # Waktu rilis dalam detik

# Terapkan noise gate
filtered_audio = noise_gate(audio_data.astype(np.float32), threshold, ratio, attack_time, release_time, fs)

# Simpan audio yang telah difilter
output_path = 'filtered_audio_noise_gate.wav'  # Ganti dengan path output yang diinginkan
with wave.open(output_path, 'w') as wav_file:
    wav_file.setnchannels(n_channels)
    wav_file.setsampwidth(sample_width)
    wav_file.setframerate(fs)
    wav_file.writeframes(filtered_audio.astype(np.int16).tobytes())

print("Audio telah diproses dengan noise gate dan disimpan sebagai:", output_path)

# Visualisasikan audio sebelum dan sesudah noise gate
plt.figure(figsize=(14, 6))
plt.subplot(2, 1, 1)
plt.plot(audio_data, label='Original Audio')
plt.title('Audio Sebelum Noise Gate')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(filtered_audio, label='Audio Setelah Noise Gate', color='orange')
plt.title('Audio Setelah Noise Gate')
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
````

**PERCAKAPAN DENGAN AI-LLM**

Me: Bagaimana cara membuat noise gate pada audio?
AI-LLM: Noise gate adalah alat yang digunakan dalam pengolahan audio untuk mengurangi suara yang berada di bawah ambang batas tertentu, yang membantu menghilangkan noise atau suara latar belakang yang tidak diinginkan. Berikut adalah cara membuat noise gate pada audio:

### Langkah-Langkah Membuat Noise Gate

1. **Tentukan Ambang Batas (Threshold):**
   - Ini adalah level di mana suara dianggap cukup keras untuk diteruskan. Suara yang di bawah level ini akan dibungkam. Misalnya, Anda dapat menetapkan threshold pada -30 dB.

2. **Atur Rasio (Ratio):**
   - Rasio menentukan seberapa banyak sinyal di atas threshold akan dilewatkan. Sebagai contoh, jika rasio ditetapkan ke 4:1, maka untuk setiap 4 dB di atas threshold, hanya 1 dB yang akan dilewatkan.

3. **Atur Waktu Serangan (Attack Time):**
   - Ini adalah waktu yang dibutuhkan untuk noise gate untuk membuka (mengizinkan suara lewat) setelah suara melewati threshold. Attack time biasanya diukur dalam milidetik.

4. **Atur Waktu Rilis (Release Time):**
   - Ini adalah waktu yang dibutuhkan untuk noise gate untuk menutup setelah suara jatuh di bawah threshold. Release time juga diukur dalam milidetik.

5. **Implementasi:**
   - Buat fungsi yang akan memproses audio dengan membandingkan level suara dengan threshold. Suara di atas threshold akan diteruskan, sementara suara di bawah threshold akan dibungkam atau dikurangi sesuai dengan rasio.

### Contoh Pseudocode Noise Gate

Berikut adalah contoh pseudocode untuk noise gate:

```plaintext
function noise_gate(audio_data, threshold, ratio, attack_time, release_time):
    output = empty array of same size as audio_data
    gain = 1.0  # Initial gain

    for each sample in audio_data:
        if sample > threshold:
            gain = 1.0  # Suara di atas threshold, tidak ada pengurangan
        else:
            gain = 0.0  # Suara di bawah threshold, dibungkam

        output[i] = sample * gain

        # Terapkan attack dan release
        if gain > 0:
            increase gain towards 1.0 over attack_time
        else:
            decrease gain towards 0 over release_time

    return output
```

```python
output_path = 'filtered_audio.wav'  # Ganti dengan path output yang diinginkan
with wave.open(output_path, 'w') as wav_file:
    wav_file.setnchannels(n_channels)
    wav_file.setsampwidth(sample_width)
    wav_file.setframerate(fs)
    wav_file.writeframes(filtered_audio.astype(np.int16).tobytes())

print("Audio telah diproses dengan noise gate dan disimpan sebagai:", output_path)
```

- **noise_gate**: Menerapkan noise gate berdasarkan threshold, rasio, waktu serangan, dan waktu rilis.
